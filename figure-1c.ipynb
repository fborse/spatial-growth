```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
from itertools import product
```

# Loading the data

All we load here are the $N_i(t)$, which are provided as a NumPy array :

```python
Nt = np.load("curves_raw.npy")
```

# Figure

```python
fig, axes = plt.subplots(ncols = 4, figsize = (20, 4))
cmap = sns.color_palette("mako_r", as_cmap = True)

for p, ax in enumerate(axes):
    sns.heatmap(Nt[p, ..., -1].T, ax = ax, cmap = cmap)
    ax.set_title(f"plate {p+1}")
    ax.set_ylabel("row")
    ax.set_xlabel("column")

fig.savefig(f"plots/figure-1/t-final.pdf", bbox_inches = "tight")
fig.savefig(f"plots/figure-1/t-final.png", bbox_inches = "tight");
```
