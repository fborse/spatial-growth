```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
from itertools import product
```

```python
fontsize = 14
```

# Loading the data

All we load here are the $N_i(t)$, which are provided as a NumPy array :

```python
Nt = np.load("curves_raw.npy")
```

# Figure

```python
fig, axes = plt.subplots(ncols = 4, figsize = (20, 4), gridspec_kw = { "width_ratios": [1, 1, 1, 1.25] })
cmap = sns.color_palette("mako_r", as_cmap = True)

envs = ["Gal + NaCl", "Glc", "Glc + NaCl", "Gal"]
for p, ax in enumerate(axes):
    sns.heatmap(Nt[p, ..., -1].T / Nt[p, ..., -1].max(), ax = ax, cmap = cmap, vmin = 0, vmax = 1, cbar = (p == 3))
    ax.set_title(envs[p], fontsize = fontsize)
    ax.set_ylabel("row", fontsize = fontsize)
    ax.set_xlabel("column", fontsize = fontsize)

fig.savefig(f"plots/figure-1/t-final.pdf", bbox_inches = "tight")
fig.savefig(f"plots/figure-1/t-final.png", bbox_inches = "tight");
```
